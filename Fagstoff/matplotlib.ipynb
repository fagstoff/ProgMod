Listene over danner punktene (0,0), (1,1), (2,4), (3,9) og (4,16) som så blir plottet ogforbundet med en rett linje.

```python
# Importerer nødvendige biblioteker
import matplotlib.pyplot as plt

# Lager to lister (tupler) med tall.
# Hvert tallpar danner et punkt.
x = (0, 1, 2, 3, 4)
y = (0, 1, 4, 9, 16)

# Utskrift av data
plt.plot(x, y)
plt.show()
```

Som regel ønsker vi mer informasjon i plottet. I eksempelet under ser vi hvordan vi kan legge til mer informasjon.

```python
z = (0, 1, 8, 27, 64)

plt.grid() # Lager rutenett
plt.xlabel('$x$') # Merker x-aksen
plt.ylabel('$f(x)$') # Merker y-aksen

plt.plot(x, y, 'bx--', label='$f(x)=x^2$') # bx-- -> blå stiplet linje med x som punktmarkering.
plt.plot(x, z, 'gt--', label='$f(x)=x^3$')
plt.legend() # Infoboksen øverst i hjørnet. Bruker data fra label
plt.show()
```

```python
I eksempelet under benyttes bibliotekene [matplotlib](http://matplotlib.org/) og [numpy](http://www.numpy.org/) til å skrive ut funksjonen: 

$$s(t) = 1 + \text{sin}(2\pi t) \text{ for alle }x\in \{0.0 , 2.0\}$$

```

```python
# Importerer nødvendige biblioteker
import matplotlib.pyplot as plt
import numpy as np

# Funksjonsdata
t = np.arange(0.0, 2.0, 0.01) # Angir definisjonsmengden til funksjonen - Punkter for hver 0.01
s = 1 + np.sin(2*np.pi*t) # Funksjonen som skal plottes
plt.plot(t, s, 'r--', label='$s(t) = 1 + \text{sin}(2\pi t) \text{ for alle }x\in \{0.0 , 2.0\}$') # Plotter s(t)

# Koden under beskriver hvordan koordinatsystemet skal se ut
plt.xlabel('tid (s)') # Navn på x-akse
plt.ylabel('spenninge (mV)') # Navn på y-akse
plt.title('En sinuskurve') # Tittel på grafen
plt.grid(True) # Rutenett
plt.savefig("test.png") # Filnavn på bildet som lagres
plt.show() # Vis bildet
```
